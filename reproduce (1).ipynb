```python
import splitfolders  # or import split_folders

# Split with a ratio.
# To only split into training and validation set, set a tuple to `ratio`, i.e, `(.8, .2)`.
splitfolders.ratio("DLAI3_Phase3/DLAI3_Phase3/", output="output", seed=1337, ratio=(.8, .2), group_prefix=None) # default values

# Split val/test with a fixed number of items e.g. 100 for each set.
# To only split into training and validation set, use a single number to `fixed`, i.e., `10`.
splitfolders.fixed("DLAI3_Phase3/DLAI3_Phase3/", output="output", seed=1337, fixed=100, oversample=False, group_prefix=None) # default values
```
